# 2 Datatyper

## Numeriske:

Variabler som fornuftig kan representeres i form av tall på datamaskinen (integre eller flyttall).

### Kontinuerlige:

Variabler som kan plasseres på en kontinuerlig skala, der intervallene kan være så små en måtte ønske. Eksempler: Lengde, amplityde, temperatur.

### Diskrete:

Variabler som det kun er meningsfylt å representere ved heltall eller naturlige tall. Eksempler: Antall stoler i et rom, patroner i et magasin, medarbeidere på en arbeidsplass.

## Kategoriske:

Variabler hvis funksjon i hovedsak er å betegne noe eller kjennetegne identiteten til et objekt. Har ikke numerisk verdi - dersom en kategorisk variabel utgjør et tall, kan ikke disse regnes med slik som numeriske variabler.

### Nominelle

Alle nøkterne 'etiketter' som gir et objekt en eller annen form for identitet. Hvis disse utgjør numre, kan de ikke inngå i matematiske operasjoner. Eksempler: Telefonnummer, personnummer, registreringsnummer på kjøretøy.

### Ordinelle

Sett med variabler som står i et visst rangforhold til hverandre / rangert høyere eller lavere enn hverandre, men der avstanden mellom hver "rang" ikke er definert / relevant. Eksempler: Universitetsgrader, stillingstitler, ranking (målgangsnummer i konkurranse / plass på sportstabell).

# 3 Supernova-data

```python
import pandas as pd
import matplotlib.pyplot as plt

df_supernova = pd.DataFrame(pd.read_csv("SN_list-1.csv", header=0))
```

```python
x = df_supernova['Type'].unique()
y = df_supernova['Type'].value_counts()

plt.bar(x, y, color="grey", edgecolor="black", linewidth=1)
for value in range(len(x)):
    plt.text(value - 0.2, y.iloc[value] + 0.2, f'{y.iloc[value]}')
plt.xlabel("Supernova Types")
plt.ylabel("Observations")
plt.title(f"Observations of supernova types\n{df_supernova['Date'].min()} - {df_supernova['Date'].max()}", fontweight="bold")
plt.show()
```
